Tracé des résultats
===================

```python
import numpy as np
import matplotlib.pyplot as plt
```

On récupère les données contenues dans le fichier .txt.

```python
def extraction (file_name):
    L=[]
    val = []
    f = open(file_name, "r")
    for line in f : 
        L.append(line.split(";"))
    m = len(L[0]) #L[i] same size
    for j in range (len(L)) :
        l = []
        for k in range (m-1) :
            l.append(float(L[j][k]))
        val.extend([l])
    return val
```

Pour vérifier que la fonction convertit bien le texte en liste de flottants: 

```python
extraction("valeurs.txt")
```

Puis on trace l'évolution de la température en chaque point.
Voici le code et le résultat avec la méthode d'Euler explicite :

```python
xf = 1.
temp_val = extraction("valeurs6.txt")
Nt = len(temp_val)
les_x = np.linspace(0.,1.,num=20, endpoint=True)
n=0
for i in range (Nt) :
    les_y = np.array(temp_val[i])
    plt.plot(les_x, les_y)
    n+=1
plt.xlabel("abscisse x")
plt.ylabel("Température")
plt.title("Résolution par la méthode d'Euler explicite")
plt.show()
```
